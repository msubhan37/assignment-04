1. Write a Python program to create a class representing a Circle. Include methods to calculate its area and perimeter.

```python
import math

class circle:
    def __init__(self,radius):
        
        self.radius=radius

    def calculate_area(self):
        return math.pi * (self.radius**2)
    
    def calculate_perimeter(self):
        return 2 * math.pi*self.radius
    
my_circle = circle(5)

print("RADIUS:", my_circle.radius)
print("AREA:", round(my_circle.calculate_area(), 2))
print("PERIMETER:", round(my_circle.calculate_perimeter(), 2))
```

2. Write a Python program to create a person class. Include attributes like name, country and date of birth. Implement a method to determine the person's age.


```python
from datetime import date

class person ():
    def __init__(self,name,country,date_of_birth):
        self.name=name
        self.country=country
        self.date_of_birth=date_of_birth

    def calculate_age(self):
        today = date.today()

        age = today.year - self.date_of_birth.year

        return age

        if (today.month, today.day) < (self.date_of_birth.month, self.date_of_birth.day):
            age -=1

person1 = person("SUBHAN", "PAKISTAN", date(2007,8,15))

print("NAME:", {person1.name})
print("COUNTRY:", {person1.country})
print("AGE:", person1.calculate_age(), "year")
```

3. Write a Python program to create a calculator class. Include methods for basic arithmetic operations.

```python
class calculator:
    def add(self, a, b):
        return a + b

    def subtract(self, a, b):
        return a - b 

    def multiply(self, a, b):
        return a * b

    def divide(self, a, b):
        if b==0:
            return("ERROR: YOU CAN NOT DIVIDED BY ZERO")
        return a / b

calc = calculator()

print("ADD (10+5):", calc.add(10,5))
print("SUB (10-5):", calc.subtract(10,5))
print("MULTIPLY (10*5):", calc.multiply(10,5))
print("DIVIDED (10/2):", calc.divide(10,2))

print("DIVISION BY ZERO:",calc.divide(10,0))
```

12. Perform the following operations on a list:
- Append one item
- Insert one item
- Remove one item
- Pop one item
- Print the list after each operation.

```python
fruits = ["Apple", "Banana", "Cherry"]
print("Initial List:      ", fruits)

fruits.append("Mango")
print("After Append:      ", fruits)

fruits.insert(1, "Grapes")
print("After Insert:      ", fruits)

fruits.remove("Banana")
print("After Remove:      ", fruits)

popped_item = fruits.pop()
print("After Pop:         ", fruits)
print("Popped Item Was:   ", popped_item)
```
